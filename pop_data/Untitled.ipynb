```python
import pandas as pd
import numpy as np
import json
from us import states
```

```python
population = pd.read_csv(open('congressional_district_population.csv'))
```

```python
population
```

```python
state_list = []
states2abbr = dict()
for state in states.STATES:
    state_list.append(state.name)
    states2abbr[state.name] = state.abbr
```

```python
state_pops = population[population['District'].isin(state_list)].reset_index(drop=True)
```

```python
def clean_states(in_df):
    df = in_df.copy()
    df = df.rename(columns={'District': 'state_po', 'Urban': 'urban', 'Suburban': 'suburban', 'Rural': 'rural', 
                            'Urban %': 'urban_pct', 'Suburban %': 'suburban_pct',
                            'Rural %': 'rural_pct', 'Most Prevalent': 'prevalent'})
    df['urban_pct'] /= 100
    df['suburban_pct'] /= 100
    df['rural_pct'] /= 100
    df['state_po'] = list(map(lambda x: states2abbr[x], state_pops['District']))
    df['prevalent'] = (df['prevalent']).str.lower()
    return df
```

```python
clean_states(state_pops)
```
