```python
import pandas as pd
```

```python
df = pd.read_stata('anes1996dta/nes96.dta')
```

```python
def clean(in_df):
    df = in_df.copy()
    df = df.drop(columns=[
        'VERSION', 'DSETID', 'ICPSRNO', 'V960113', 'V960126',
        'V960126A', 'V960594', 'V960595', 'V960596', 'V960620',
        'V960632', 'V960644', 'V960657', 'V960666', 'V960677',
        'V960695', 'V961077', 'V961086', 'V961463'
    ])
    return df
```

```python
clean(df)
```
